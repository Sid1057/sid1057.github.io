```python
from torch import nn, optim
import torch
import torchvision
```

```python
FCN = torchvision.models.segmentation.fcn_resnet101(pretrained=True)
DeepLab = torchvision.models.segmentation.deeplabv3_resnet101(pretrained=True)
RCNN = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
MaskRCNN = torchvision.models.detection.maskrcnn_resnet50_fpn(pretrained=True)
```

```python
from torchvision.models._utils import IntermediateLayerGetter
import torchvision.models.resnet as resnet

from torch.nn import functional as F

from collections import OrderedDict
from torchvision.models.segmentation.deeplabv3 import DeepLabHead

squeeze = torchvision.models.squeezenet.squeezenet1_1(pretrained=True).eval()

def tensor_preproc(tensor):
    return tensor.cuda(0).half()

class DeepSqueeze(torch.nn.Module):
    def __init__(self, num_classes):
        super(DeepSqueeze, self).__init__()
        self.backbone = squeeze.features

        self.classifier = DeepLabHead(512, num_classes)

        self.aux_classifier = None

    def forward(self, x):
        input_shape = x.shape[-2:]
        x = tensor_preproc(x)
        features = self.backbone(x)

        result = OrderedDict()

        x = self.classifier(features)
        x = F.interpolate(x, size=input_shape, mode='bilinear', align_corners=False)

        return x

model_semantic_carla = DeepSqueeze(35).eval()
model_semantic_carla = tensor_preproc(model_semantic_carla)
model_semantic_carla.load_state_dict(
    torch.load('DeepLab_v3_squeeze11_kitti_classes_iou_mean012_iou_max60.pth'), strict=False)
```

```python
models = [
    ('FCN resnet101', FCN),
    ('DeepLab resnet101', DeepLab),
    ('Faster-RCNN resnet50', RCNN),
    ('MaskRCNN resnet50', MaskRCNN),
    ('DeepLab SqueezeNet', model_semantic_carla),
]

allocated_memory_plot_dict = dict()

for _, model in models:
    model.cpu()
    
for _, input in inputs:
    input.cpu()

torch.cuda.empty_cache()

still_allocated, still_cached = 0, 0

with torch.no_grad():
    for model_name, model in models:
        model.cuda(0).eval()
        print(model_name, '\n')
        print('Pure allocated: {}mb; cached: {}mb\n'.format(
            int(torch.cuda.memory_allocated()/1024**2), int(torch.cuda.memory_cached()/1024**2)))
        allocated_memory_plot_dict.update({model_name: dict()})

        for name, input in inputs:
            allocated_memory_plot_dict[model_name].update({name: dict()})
            input.cpu()
            input = input.cuda(0)
            string = %timeit -r 10 -n 10 -q -o model(input)

            allocated = int(torch.cuda.memory_allocated()/1024**2 - still_allocated)
            cached = int(torch.cuda.memory_cached()/1024**2 - still_cached)
            mean, std = string.average, string.stdev

            print('With input {} allocated: {}mb; cached: {}mb'.format(
                name, allocated, cached))
            print('Mean time: {:.2f}(ms) std: {:.2f}(ms)\n'.format(mean*1000, std*1000))

            input.cpu()
            torch.cuda.empty_cache()

            allocated_memory_plot_dict[model_name][name].update({
                "mean time (ms)": string.average*1000, 
                "std time (ms)": string.stdev*1000,
                "best time (ms)": string.best*1000,
                "worst time (ms)": string.worst*1000,
                "allocated memory (mb)": allocated,
                "cached memory (mb)": cached,
            })

        model.cpu()
        torch.cuda.empty_cache()
        still_allocated, still_cached = torch.cuda.memory_allocated()/1024**2, torch.cuda.memory_cached()/1024**2
        print('After empty_cache method: {}mb; cached: {}mb\n\n'.format(
            still_allocated, still_cached))
```

```python
CarlaClasses_converter_index_dict = {
     0: [0,    0,   0,   0,   1,],  # None
     1: [11,   1,   1,   2,   3,],  # Buildings
     2: [13,   2,   2,   2,   3,],  # Fences
     3: [23,   3,   3,   5,   6,],  # Other   SKY??
     4: [24,   4,   4,   6,   7,],  # Pedestrians
     5: [17,   5,   5,   3,   4,],  # Poles
     6: [7,    6,   7,   1,   0,],  # RoadLines
     7: [7,    7,   7,   1,   0,],  # Roads
     8: [8,    8,   8,   1,   2,],  # Sidewalks
     9: [21,   9,   9,   4,   5,],  # Vegetation
    10: [26,  10,  10,   7,   8,],  # Vehicles
    11: [12,  11,  11,   2,   3,],  # Walls
    12: [20,  12,  12,   3,   4,],  # TrafficSigns
}
```

```python
def CityscapesTransform(type_from, type_to):
    if not type_from in ['CityScapesClasses', 'CarlaClasses']:
        raise KeyError
        
    if not type_to in [
            'CityScapesClasses',
            'CarlaClassesWithRoadLines', 'CarlaClassesWithoutRoadLines',
            'Categories', 'CategoriesWithRoad']:
        raise KeyError

    index_converter = None
    
    if type_from == 'CityScapesClasses':
        index_converter = lambda x: CityScapesClasses_converter_index_dict[x][{
            'CityScapesClasses': 0,
            'CarlaClassesWithRoadLines': 1,
            'CarlaClassesWithoutRoadLines': 2,
            'Categories': 3,
            'CategoriesWithRoad': 4
        }[type_to]]
    elif type_from == 'CarlaClasses':
        index_converter = lambda x: CarlaClasses_converter_index_dict[x][{
            'CityScapesClasses': 0,
            'CarlaClassesWithRoadLines': 1,
            'CarlaClassesWithoutRoadLines': 2,
            'Categories': 3,
            'CategoriesWithRoad': 4
        }[type_to]]
        
    def CityScapesSemanticTarget(input):
        mask = np.asarray(input).astype(np.uint8)
        target = np.zeros((
                mask.shape[0],
                mask.shape[1]
        ), dtype=np.long)

        classes_count = 35
        if type_from == 'CarlaClasses':
            classes_count = 13
            mask = mask[..., 2]
#             print(mask)

        for city_class in range(classes_count):
            channel = city_class
            target[np.where(mask == channel)] = index_converter(channel)

        return target
    
    return CityScapesSemanticTarget
```

```python
def decode_cityscapes(image, mode='CityScapesClasses'):
    if not mode in [
        'CityScapesClasses',
        'CarlaClassesWithRoadLines', 'CarlaClassesWithoutRoadLines',
            'Categories', 'CategoriesWithRoad']:
        raise KeyError

    colors = None
    max_classses = 35

    if mode == 'CityScapesClasses':
        colors = np.asarray([
            i.color for i in torchvision.datasets.Cityscapes.classes])
        max_classses = 35

    elif mode == 'CarlaClassesWithRoadLines':
        colors = np.asarray([
            (0, 0, 0),
            (70, 70, 70),
            (190, 153, 153),
            (70, 130, 180),
            (220, 20, 60),
            (153, 153, 153),
            (255, 255, 128),
            (128, 64, 128),
            (244, 35, 232),
            (107, 142, 35),
            (0, 0, 142),
            (102, 102, 156),
            (220, 220, 0)
        ])
        max_classses = 13

    elif mode == 'CarlaClassesWithoutRoadLines':
        colors = np.asarray([
            (0, 0, 0),
            (70, 70, 70),
            (190, 153, 153),
            (70, 130, 180),
            (220, 20, 60),
            (153, 153, 153),
            (128, 64, 128),
            (128, 64, 128),
            (244, 35, 232),
            (107, 142, 35),
            (0, 0, 142),
            (102, 102, 156),
            (220, 220, 0)
        ])
        max_classses = 13

    elif mode == 'Categories':
        colors = np.asarray([
            (0, 0, 0),
            (128, 64, 128),
            (70, 70, 70),
            (153, 153, 153),
            (107, 142, 35),
            (70, 130, 180),
            (220, 20, 60),
            (0, 0, 142), ])
        max_classses = 8

    elif mode == 'CategoriesWithRoad':
        colors = np.asarray([
            (128, 64, 128),
            (0, 0, 0),
            (244, 35, 232),
            (70, 70, 70),
            (153, 153, 153),
            (107, 142, 35),
            (70, 130, 180),
            (220, 20, 60),
            (0, 0, 142), ])
        max_classses = 9

    r = np.zeros_like(image).astype(np.uint8)
    g = np.zeros_like(image).astype(np.uint8)
    b = np.zeros_like(image).astype(np.uint8)

    for l in range(max_classses):
        idx = image == l
        r[idx] = colors[l, 0]
        g[idx] = colors[l, 1]
        b[idx] = colors[l, 2]

    rgb = np.stack([r, g, b], axis=2)
    return rgb
```

```python
import cv2 as cv

def IOU_error(output, target):
    output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy().astype(np.uint8)
    target = target.detach().cpu().numpy().astype(np.uint8)[0]
    diff = cv.absdiff(output, target)
    return 1.0*np.count_nonzero(diff) / diff.shape[0] / diff.shape[1]

def IOU_error(output, target):
    diff = cv.absdiff(output, target)
    return 1.0*np.count_nonzero(diff) / diff.shape[0] / diff.shape[1]
```

```python
COCO_INSTANCE_CATEGORY_NAMES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack', 'umbrella', 'N/A', 'N/A',
    'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball',
    'kite', 'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'N/A', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl',
    'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
    'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'N/A', 'dining table',
    'N/A', 'N/A', 'toilet', 'N/A', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'N/A', 'book',
    'clock', 'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]
 
def random_colour_masks(image):
    r = np.zeros_like(image).astype(np.uint8)
    g = np.zeros_like(image).astype(np.uint8)
    b = np.zeros_like(image).astype(np.uint8)
    r[image == 1], g[image == 1], b[image == 1] = 255, 255, 255
    coloured_mask = np.stack([r, g, b], axis=2)
    return r


results = []
targets = []

iou_sq_list = []
iou_fcn_list = []
iou_dl_list = []
iou_mask_list = []

model_semantic_carla = tensor_preproc(model_semantic_carla)
FCN = FCN.cuda()
DeepLab = DeepLab.cuda()
MaskRCNN = MaskRCNN.cuda()

for i, (input, target) in enumerate(CARLA_test_loader):
    def sq(input, target):
        output = model_semantic_carla(input)

        output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy()

        car_mask = cv.inRange(output, 26, 30)
#         car_mask = cv.cvtColor(car_mask, cv.COLOR_GRAY2BGR)

        car_gt = target.cpu().numpy()[0].astype(np.uint8)[0]
        car_gt = cv.inRange(car_gt, 10, 10)
#         car_gt = cv.cvtColor(car_gt, cv.COLOR_GRAY2BGR)
        print(i, 'Our IOU error: {:.2f}%'.format(IOU_error(car_gt, car_mask)*100))

        output = CityscapesTransform('CityScapesClasses', 'CarlaClassesWithoutRoadLines')(output)
        output = decode_cityscapes(output, mode='CarlaClassesWithoutRoadLines')
        
        return IOU_error(car_gt, car_mask)*100, cv.cvtColor(car_mask, cv.COLOR_GRAY2BGR)

    def fcn(input, target):
        output = FCN(input.cuda())['out']
        output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy()

        car_mask = cv.inRange(output, 7, 7)

        car_gt = target.cpu().numpy()[0].astype(np.uint8)[0]
        car_gt = cv.inRange(car_gt, 10, 10)

        print(i, 'FCN IOU error: {:.2f}%'.format(IOU_error(car_gt, car_mask)*100))
    
        return IOU_error(car_gt, car_mask)*100, cv.cvtColor(car_mask, cv.COLOR_GRAY2BGR)
    
    def dl(input, target):
        output = DeepLab(input.cuda())['out']
        output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy()

        car_mask = cv.inRange(output, 7, 7)

        car_gt = target.cpu().numpy()[0].astype(np.uint8)[0]
        car_gt = cv.inRange(car_gt, 10, 10)
        print(i, 'DL IOU error: {:.2f}%'.format(IOU_error(car_gt, car_mask)*100))

        return IOU_error(car_gt, car_mask)*100, cv.cvtColor(car_mask, cv.COLOR_GRAY2BGR)
    
    def mask(input, target):
#         output = DeepLab(input.cuda())['out']
        pred = MaskRCNN([input.cuda()[0]])
        pred_score = list(pred[0]['scores'].detach().cpu().numpy())
        pred_t = [pred_score.index(x) for x in pred_score if x>0.5]
        car_mask = np.zeros((480, 640), dtype=np.uint8)
        if len(pred_t) != 0:
            pred_t = pred_t[-1]
            masks = (pred[0]['masks']>0.5).squeeze().detach().cpu().numpy()
            pred_class = [COCO_INSTANCE_CATEGORY_NAMES[i] for i in list(pred[0]['labels'].cpu().numpy())]
            pred_boxes = [[(i[0], i[1]), (i[2], i[3])] for i in list(pred[0]['boxes'].detach().cpu().numpy())]
            masks = masks[:pred_t+1]
            pred_boxes = pred_boxes[:pred_t+1]
            pred_class = pred_class[:pred_t+1]

            for j in range(len(masks)):
                if pred_class[j] != 'car':
                    continue
                rgb_mask = random_colour_masks(masks[j])
                car_mask = cv.add(car_mask, rgb_mask)

        car_gt = target.cpu().numpy()[0].astype(np.uint8)[0]
        car_gt = cv.inRange(car_gt, 10, 10)
        print(i, 'Mask IOU error: {:.2f}%'.format(IOU_error(car_gt, car_mask)*100))
        return IOU_error(car_gt, car_mask)*100, cv.cvtColor(car_mask, cv.COLOR_GRAY2BGR)

        


    iou_sq, output_sq = sq(input, target)
    iou_fcn, output_fcn = fcn(input, target)
    iou_dl, output_dl = dl(input, target)
    iou_mask, output_mask = mask(input, target)
    
    iou_sq_list.append(iou_sq)
    iou_fcn_list.append(iou_fcn)
    iou_dl_list.append(iou_dl)
    iou_mask_list.append(iou_mask)

    target = target.type(torch.LongTensor)[0][0]
    target = decode_cityscapes(target, mode='CarlaClassesWithoutRoadLines')
    
    results.append((np.transpose((input.cpu().numpy()[0]*255).astype(np.uint8), (1, 2, 0)), output_sq, output_fcn, output_dl, output_mask, target))
#     break
```

### Классификацию

- Необходимо распознать общий класс окружающий среды для выбора оптимальный претренированных сетей

![](./cond.jpg)

### Детекцию объектов

- Необходимо детектировать сам факт присутствия объекта (дор. знаки)
- Искомый объект всегда подходит под прямоугольную форму (можно комплексировать)
- Точность комплексирования не важна
- Нужен трекинг объекта

![](./znak.jpg) 

### Сегментацию

- Необходимо использовать sensor fusion с датчиками, дающими информацию о 3D пространстве
- Необходимо находить объекты большой площади на кадре
- Разделение на сущности не важно

![](./road.jpeg) 

### Объектную сегментацию

- Необходимо получать данные с высокой достоверностью
- Искомые объекты можно разделить на сущности
- Позволяют аппаратные ресурсы
- Нужен трекинг объекта

![](./ins.png)
