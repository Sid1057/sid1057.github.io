```python
CarlaClasses_converter_index_dict = {
     0: [0,    0,   0,   0,   1,],  # None
     1: [11,   1,   1,   2,   3,],  # Buildings
     2: [13,   2,   2,   2,   3,],  # Fences
     3: [23,   3,   3,   5,   6,],  # Other   SKY??
     4: [24,   4,   4,   6,   7,],  # Pedestrians
     5: [17,   5,   5,   3,   4,],  # Poles
     6: [7,    6,   7,   1,   0,],  # RoadLines
     7: [7,    7,   7,   1,   0,],  # Roads
     8: [8,    8,   8,   1,   2,],  # Sidewalks
     9: [21,   9,   9,   4,   5,],  # Vegetation
    10: [26,  10,  10,   7,   8,],  # Vehicles
    11: [12,  11,  11,   2,   3,],  # Walls
    12: [20,  12,  12,   3,   4,],  # TrafficSigns
}
```

```python
def CityscapesTransform(type_from, type_to):
    if not type_from in ['CityScapesClasses', 'CarlaClasses']:
        raise KeyError
        
    if not type_to in [
            'CityScapesClasses',
            'CarlaClassesWithRoadLines', 'CarlaClassesWithoutRoadLines',
            'Categories', 'CategoriesWithRoad']:
        raise KeyError

    index_converter = None
    
    if type_from == 'CityScapesClasses':
        index_converter = lambda x: CityScapesClasses_converter_index_dict[x][{
            'CityScapesClasses': 0,
            'CarlaClassesWithRoadLines': 1,
            'CarlaClassesWithoutRoadLines': 2,
            'Categories': 3,
            'CategoriesWithRoad': 4
        }[type_to]]
    elif type_from == 'CarlaClasses':
        index_converter = lambda x: CarlaClasses_converter_index_dict[x][{
            'CityScapesClasses': 0,
            'CarlaClassesWithRoadLines': 1,
            'CarlaClassesWithoutRoadLines': 2,
            'Categories': 3,
            'CategoriesWithRoad': 4
        }[type_to]]
        
    def CityScapesSemanticTarget(input):
        mask = np.asarray(input).astype(np.uint8)
        target = np.zeros((
                mask.shape[0],
                mask.shape[1]
        ), dtype=np.long)

        classes_count = 35
        if type_from == 'CarlaClasses':
            classes_count = 13
            mask = mask[..., 0]

        for city_class in range(classes_count):
            channel = city_class
            target[np.where(mask == channel)] = index_converter(channel)

        return target
    
    return CityScapesSemanticTarget
```

### Create targets visualization

```python
def decode_cityscapes(image, mode='CityScapesClasses'):
    if not mode in [
        'CityScapesClasses',
        'CarlaClassesWithRoadLines', 'CarlaClassesWithoutRoadLines',
            'Categories', 'CategoriesWithRoad']:
        raise KeyError

    colors = None
    max_classses = 35

    if mode == 'CityScapesClasses':
        colors = np.asarray([
            i.color for i in torchvision.datasets.Cityscapes.classes])
        max_classses = 35

    elif mode == 'CarlaClassesWithRoadLines':
        colors = np.asarray([
            (0, 0, 0),
            (70, 70, 70),
            (190, 153, 153),
            (70, 130, 180),
            (220, 20, 60),
            (153, 153, 153),
            (255, 255, 128),
            (128, 64, 128),
            (244, 35, 232),
            (107, 142, 35),
            (0, 0, 142),
            (102, 102, 156),
            (220, 220, 0)
        ])
        max_classses = 13

    elif mode == 'CarlaClassesWithoutRoadLines':
        colors = np.asarray([
            (0, 0, 0),
            (70, 70, 70),
            (190, 153, 153),
            (70, 130, 180),
            (220, 20, 60),
            (153, 153, 153),
            (128, 64, 128),
            (128, 64, 128),
            (244, 35, 232),
            (107, 142, 35),
            (0, 0, 142),
            (102, 102, 156),
            (220, 220, 0)
        ])
        max_classses = 13

    elif mode == 'Categories':
        colors = np.asarray([
            (0, 0, 0),
            (128, 64, 128),
            (70, 70, 70),
            (153, 153, 153),
            (107, 142, 35),
            (70, 130, 180),
            (220, 20, 60),
            (0, 0, 142), ])
        max_classses = 8

    elif mode == 'CategoriesWithRoad':
        colors = np.asarray([
            (128, 64, 128),
            (0, 0, 0),
            (244, 35, 232),
            (70, 70, 70),
            (153, 153, 153),
            (107, 142, 35),
            (70, 130, 180),
            (220, 20, 60),
            (0, 0, 142), ])
        max_classses = 9

    r = np.zeros_like(image).astype(np.uint8)
    g = np.zeros_like(image).astype(np.uint8)
    b = np.zeros_like(image).astype(np.uint8)

    for l in range(max_classses):
        idx = image == l
        r[idx] = colors[l, 0]
        g[idx] = colors[l, 1]
        b[idx] = colors[l, 2]

    rgb = np.stack([r, g, b], axis=2)
    return rgb
```

```python
import torchvision.transforms as T

def torch_unsqueeze(input):
    return input.unsqueeze(0)

input_transform = T.Compose([
    #T.ToPILImage(),  # uncomment for usage with opencv images
    T.Resize(480, interpolation=PIL.Image.NEAREST),
    T.CenterCrop((480, 640)),
    T.ToTensor(), 
])

def target_transform(type_in, type_out):
    return T.Compose([
        #T.ToPILImage(),  # uncomment for usage with opencv images
    #     T.Resize(480),
        T.Resize(480, interpolation=PIL.Image.NEAREST),
        T.CenterCrop((480, 640)),
        CityscapesTransform(type_in, type_out),
        T.ToTensor(),
    ])

def target_in_net_transform(x):
    return x.type(torch.LongTensor)[0]
```

## Fit several models with different targets

### Semantic segmentation model

#### Architecture

```python
from torchvision.models._utils import IntermediateLayerGetter
import torchvision.models.resnet as resnet

from torch.nn import functional as F

from collections import OrderedDict
from torchvision.models.segmentation.deeplabv3 import DeepLabHead

squeeze = torchvision.models.squeezenet.squeezenet1_1(pretrained=True).eval()

class DeepSqueeze(torch.nn.Module):
    def __init__(self, num_classes):
        super(DeepSqueeze, self).__init__()
        self.backbone = squeeze.features

        self.classifier = DeepLabHead(512, num_classes)

        self.aux_classifier = None

    def forward(self, x):
        input_shape = x.shape[-2:]
        features = self.backbone(x)

        result = OrderedDict()

        x = self.classifier(features)
        x = F.interpolate(x, size=input_shape, mode='bilinear', align_corners=False)

        return x
```

#### Loss function and lists for history

```python
loss_fn = torch.nn.CrossEntropyLoss().cuda(2)
```

```python
def IOU_error(output, target):
    output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy().astype(np.uint8)
    target = target.detach().cpu().numpy().astype(np.uint8)[0]
    diff = cv.absdiff(output, target)
    return 1.0*np.count_nonzero(diff) / diff.shape[0] / diff.shape[1]
```

### Fitting

```python
# torch.save(model.state_dict(), 'DeepLab_v3_squeeze11_kitti_classes_loss_mean041_loss_max055_iou_max72.pth')
model_semantic_carla.load_state_dict(
    torch.load('DeepLab_v3_squeeze11_kitti_classes_iou_mean012_iou_max60.pth'), strict=False)
# torch.save(
#     model_semantic_carla.state_dict(),
#     'DeepLab_v3_squeeze11_kitti_classes_iou_mean012_iou_max60.pth')
```

```python
import time
import random

model = model_semantic_carla

train_loader = CityScapes_train_loader
val_loader = CityScapes_val_loader

# train_loader = KITTI_train_loader
# val_loader = KITTI_val_loader

# train_loader = KITTI_val_loader
# val_loader = KITTI_train_loader

thresholdes = [0.9, 0.8, 0.7, 0.6, 0.55]
early_stopping = False

def validate_closure(input, target):
    optimizer.zero_grad()
    input = tensor_preproc(input)
    output = model(input).detach()
    target = tensor_preproc(target[0])
#     print(type(target))
    loss = IOU_error(output.type(torch.FloatTensor), target.type(torch.LongTensor))
    optimizer.zero_grad()
    return loss, output


def closure(input, target):
    optimizer.zero_grad()
    input = tensor_preproc(input)
    output = model(input)
    target = tensor_preproc(target[0])
    loss = loss_fn(output.type(torch.FloatTensor).cuda(2), target.type(torch.LongTensor).cuda(2))
    loss.backward()
    return loss, output


def test_closure(input, target):
    optimizer.zero_grad()
    input = tensor_preproc(input)
    output = model(input)
    target = tensor_preproc(target[0])
    loss = loss_fn(output.type(torch.FloatTensor).cuda(2), target.type(torch.LongTensor).cuda(2))
    return loss, output
        

matplotlib.rcParams['figure.figsize'] = [9.8, 9.6]

for iteration in range(3):
    if early_stopping:
        break

    optimizer = torch.optim.SGD([p for p in model.parameters() if p.requires_grad], lr=1e-2)
    scheduler_val = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, patience=2, verbose=True, factor=0.4)
    scheduler_val_max = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, patience=3, verbose=True, factor=0.5, eps=1e-4-1e-5+1e-6)
    print('\nIteration {}\n'.format(iteration))
    

    for loss_threshold_index, loss_threshold in enumerate(thresholdes):
        if early_stopping:
            break

        print('\nCurrent loss threshold:', loss_threshold)

        scheduler_threshold = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, patience=2, verbose=True, factor=0.5, eps=1e-4-1e-5+1e-6)

        for epoch in range(10):
            torch.cuda.empty_cache()
            loss_mean = 0
            val_loss_mean = 0

            print('Epoch: ', epoch)
            t = time.time()

            bad_pictures = []
            for i, (input, target) in enumerate(train_loader):
                loss = test_closure(input, target)[0]
                if loss.item() > loss_threshold:
                    bad_pictures.append(i)
                    loss = closure(input, target)[0]
                    optimizer.step()

                loss_mean += loss.item()
                print(' '*45, end='\r')
                print(
                    'Train progress: {:.02f}% ({}/{})'.format(
                        100*(i+1)/len(train_loader), i+1, len(train_loader)),
                    '; {} worse then {}'.format(
                        len(bad_pictures), loss_threshold), end='\r')
            print()
            torch.cuda.empty_cache()

            if len(bad_pictures) == 0:
                if loss_threshold_index == len(thresholdes)-1:
                    print('Early stopping')
                    early_stopping = True
                    torch.cuda.empty_cache()
                break

            torch.cuda.empty_cache()

            val_max = 0
            for i, (input, target) in enumerate(val_loader):
                loss, output = validate_closure(input, target)
                val_max = max(val_max, loss)
                val_loss_mean += loss
                print(' '*45, end='\r')
                print('Validation progress: {:.02f}'.format(100*(i+1)/len(val_loader)), '%', end='\r')
                if (epoch+1) % 5 == 0 and i == len(val_loader)-1:
                    output = torch.argmax(output.squeeze(), dim=0).detach().cpu().numpy()
                    output = decode_cityscapes(output, mode='CityScapesClasses')

                    target = target.type(torch.LongTensor)[0][0]
                    target = decode_cityscapes(target, mode='CityScapesClasses')

                    input_original = np.transpose(input.cpu().numpy()[0]*255, (1, 2, 0)).astype(np.uint8)
                    input_output = cv.addWeighted(input_original, 1, output, 0.5, 0)
                    input_target = cv.addWeighted(input_original, 1, target, 0.5, 0)
                    error = cv.absdiff(cv.cvtColor(output, cv.COLOR_BGR2GRAY), cv.cvtColor(target, cv.COLOR_BGR2GRAY))
                    input_error = input_original.copy()
                    input_error[np.where(error > 0)] = (255, 0, 0)

                    plt.title('Model output example')
                    plt.imshow(
                        np.vstack((
                            np.hstack((
                                input_original,
                                input_error)),
                            np.hstack((
                                input_target,
                                target)),
                            np.hstack((
                                input_output,
                                output)),
                        )))
                    plt.show()
                    
                    plt.title('Validation and train loss')
#                     matplotlib.rcParams['figure.figsize'] = [12.8/2, 9.6/2]
                    plt.plot(val_story, 'r-')
                    plt.plot(val_story_max, 'g.')
                    plt.plot(story, 'b-')
                    plt.show()

            torch.cuda.empty_cache()

            story.append(loss_mean/len(train_loader))
            val_story.append(val_loss_mean/len(val_loader))
            val_story_max.append(val_max)
            print('Loss: ', story[-1], '; val:', val_story[-1], '; val max:', val_max)
            print('Time per epoch: {:.2f}(s)'.format(time.time()-t))

            scheduler_threshold.step(len(bad_pictures))
            if loss_threshold_index > 0:
                scheduler_val.step(val_loss_mean/len(val_loader))
            scheduler_val_max.step(val_max)

#             torch.save(
#                 net_test.state_dict(),
#                 'DeepLab_v3_resnet18_cityscapes_category_{}_{}.pth'.format(
#                     loss_threshold_index, epoch))

    torch.save(
        model.state_dict(),
        'DSQ_CS_iteration_{}.pth'.format(iteration))
```
